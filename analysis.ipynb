# Student Sleep & Productivity Analyzer

## Objective
Analyze how sleep, screen time, and caffeine intake relate to student productivity.

## Dataset
Synthetic student lifestyle data generated to reflect realistic behavior ranges.

## Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(x="sleep_hours", y="productivity_score", data=df)
plt.title("Sleep Hours vs Productivity")
plt.xlabel("Sleep Hours")
plt.ylabel("Productivity Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="screen_time_hours", y="productivity_score", data=df)
plt.title("Screen Time vs Productivity")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Productivity Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="caffeine_mg", y="sleep_hours", data=df)
plt.title("Caffeine Intake vs Sleep Duration")
plt.xlabel("Caffeine (mg)")
plt.ylabel("Sleep Hours")
plt.show()
```

```python
df[[
    "sleep_hours",
    "screen_time_hours",
    "caffeine_mg",
    "study_hours",
    "productivity_score"
]].corr()
```

## Key Insights
- Sleep duration shows a positive correlation with productivity
- Screen time exhibits a mild negative relationship with productivity
- Caffeine intake does not strongly correlate with improved sleep or productivity
